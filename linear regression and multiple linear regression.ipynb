```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

n = len(X)
k = 1


m, c = np.polyfit(X, y, 1)


y_pred = m * X + c

print(f"Regression Line: y = {m:.2f}x + {c:.2f}\n")




mse = mean_squared_error(y, y_pred)
print(f"MSE (Mean Squared Error): {mse:.4f}")


rmse = np.sqrt(mse)
print(f"RMSE (Root Mean Squared Error): {rmse:.4f}")

mae = mean_absolute_error(y, y_pred)
print(f"MAE (Mean Absolute Error): {mae:.4f}")


r_squared = r2_score(y, y_pred)
print(f"R-squared (R2): {r_squared:.4f}")

if (n - k - 1) > 0:
    adj_r_squared = 1 - (1 - r_squared) * (n - 1) / (n - k - 1)
    print(f"Adjusted R-squared (Adj R2): {adj_r_squared:.4f}")
else:
    print("Adjusted R-squared cannot be calculated for this data (n - k - 1 <= 0)")


if (n - k - 1) > 0:
    see = np.sqrt(np.sum((y - y_pred)**2) / (n - k - 1))
    print(f"SEE (Standard Error of the Estimate): {see:.4f}")
else:
    print("SEE cannot be calculated for this data (n - k - 1 <= 0)")

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label=f'Regression Line (y = {m:.2f}x + {c:.2f})')
plt.title('Simple Linear Regression with Performance Metrics')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

X1 = np.array([1,1,2])
X2 = np.array([1, 2,2])
Y = np.array([6, 8, 9])


X = np.column_stack((X1, X2))

print(f"X1 array: {X1}")
print(f"X2 array: {X2}")
print(f"Y array: {Y}")
print(f"Combined feature matrix X:\n{X}")
print(f"Shape of X: {X.shape}")
```

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression()

model.fit(X, Y)
y_pred = model.predict(X)

m_coeffs = model.coef_
c_intercept = model.intercept_

print(f"Predicted Y values: {y_pred}")
print(f"Coefficients (m_coeffs): {m_coeffs}")
print(f"Intercept (c_intercept): {c_intercept}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(Y, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.4f}")


mse = mean_squared_error(Y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.4f}")


rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")


r_squared = r2_score(Y, y_pred)
print(f"R-squared (R2): {r_squared:.4f}")

n = len(Y)
k = X.shape[1]


if (n - k - 1) > 0:
    adj_r_squared = 1 - (1 - r_squared) * (n - 1) / (n - k - 1)
    print(f"Adjusted R-squared (Adj R2): {adj_r_squared:.4f}")
else:
    print("Adjusted R-squared cannot be calculated for this data (n - k - 1 <= 0)")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X1, X2, Y, color='blue', label='Original Data Points', s=50)

x1_surf = np.linspace(X1.min() - 0.5, X1.max() + 0.5, 10)
x2_surf = np.linspace(X2.min() - 0.5, X2.max() + 0.5, 10)

X1_mesh, X2_mesh = np.meshgrid(x1_surf, x2_surf)

Y_plane = c_intercept + m_coeffs[0] * X1_mesh + m_coeffs[1] * X2_mesh

ax.plot_surface(X1_mesh, X2_mesh, Y_plane, color='red', alpha=0.5, label='Regression Plane')

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

ax.set_title('Multiple Linear Regression: 3D Plane')

ax.scatter([], [], [], color='red', alpha=0.5, label='Regression Plane')

ax.legend()

plt.show()
```
