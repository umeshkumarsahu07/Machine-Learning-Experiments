```python


import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

X = np.array([[2], [4], [0], [1]])
y = np.array([1, 1, -1, -1])


model = SVC(kernel='linear', C=1e6)
model.fit(X, y)


w = model.coef_[0][0]
b = model.intercept_[0]

print("SVM Model:")
print("Weight (w) =", w)
print("Bias (b) =", b)

print("\nDecision Function:")
print(f"f(x) = {w:.2f}x + ({b:.2f})")

print("\nDecision Boundary:")
print(f"x = {-b/w:.2f}")


x_test = np.array([[3]])
prediction = model.predict(x_test)

print("\nTest Point x = 3")
print("Predicted Class =", prediction[0])


plt.figure(figsize=(10, 5))


plt.scatter(X[y==-1], y[y==-1],
            color='red', s=120, label='Class -1')


plt.scatter(X[y==1], y[y==1],
            color='blue', s=120, label='Class +1')


sv = model.support_vectors_
plt.scatter(sv, model.predict(sv),
            s=250, facecolors='none',
            edgecolors='black', linewidths=2,
            label='Support Vectors')


x_boundary = -b / w
plt.axvline(x=x_boundary, color='green',
            linestyle='--', linewidth=2,
            label='Decision Boundary')


plt.axvline(x=1, color='gray', linestyle=':')
plt.axvline(x=2, color='gray', linestyle=':')


plt.scatter(3, prediction[0],
            color='purple', marker='x',
            s=200, linewidths=3,
            label='Test Point (x=3)')


plt.yticks([-1, 1])
plt.ylim(-2, 2)
plt.xlabel("Feature x")
plt.ylabel("Class Label")
plt.title("1D Support Vector Machine Visualization")
plt.legend()
plt.grid(True)

plt.show()
```

```python



X = np.array([
    [3, 3],
    [4, 3],
    [1, 1],
    [2, 1]
])

y = np.array([1, 1, -1, -1])


model = SVC(kernel='linear', C=1e6)
model.fit(X, y)


w = model.coef_[0]
b = model.intercept_[0]

print("SVM Model Parameters:")
print("w1 =", w[0])
print("w2 =", w[1])
print("b  =", b)

print("\nDecision Function:")
print(f"f(x) = {w[0]:.3f}x1 + {w[1]:.3f}x2 + ({b:.3f})")


x_test = np.array([[3, 3]])
prediction = model.predict(x_test)

print("\nTest Point (3,3)")
print("Predicted Class =", prediction[0])


plt.figure(figsize=(8,6))


plt.scatter(X[y==1][:,0], X[y==1][:,1],
            color='blue', s=120, label='Class +1')
plt.scatter(X[y==-1][:,0], X[y==-1][:,1],
            color='red', s=120, label='Class -1')


sv = model.support_vectors_
plt.scatter(sv[:,0], sv[:,1],
            s=250, facecolors='none',
            edgecolors='black', linewidths=2,
            label='Support Vectors')


x_vals = np.linspace(-1, 5, 100)
y_vals = -(w[0]*x_vals + b) / w[1]
plt.plot(x_vals, y_vals, 'g--', label='Decision Boundary')


margin = 1 / np.linalg.norm(w)
y_margin_up = -(w[0]*x_vals + b - 1) / w[1]
y_margin_down = -(w[0]*x_vals + b + 1) / w[1]

plt.plot(x_vals, y_margin_up, 'k:')
plt.plot(x_vals, y_margin_down, 'k:')


plt.scatter(3, 3, color='purple',
            marker='x', s=200, linewidths=3,
            label='Test Point (3,3)')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("2D Linear SVM Classification")
plt.legend()
plt.grid(True)

plt.show()
```
